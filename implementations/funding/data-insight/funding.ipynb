````python
import json
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

# Load metric model meta data
cout_org_datasource = json.load(open('../data/cout_org_datasource.json'))
org_contributors = json.load(open('../data/contributors_of_org.json'))


# Draw a column diagram
def draw_bar(df, y, ylabel, title):
    df['date'] = pd.to_datetime(df.grimoire_creation_date)
    df.sort_values('date', inplace=True)
    fig, ax = plt.subplots(figsize=(30, 6));
    fig = sns.barplot(x='date', y=y, data=df, ci=None, ax=ax, palette="Blues")
    ax.set(xlabel="Year-Month-Day", ylabel=ylabel, title=title)
    x_dates = df['date'].dt.strftime('%Y-%m-%d').unique()
    ax.set_xticklabels(labels=x_dates, rotation=45, ha='right');
````

### 1. Organizational Diversity
Description

Organizational diversity expresses how many different organizations are involved in a project and how involved different organizations are compared to one another.


````python
import matplotlib.pyplot as plt
import matplotlib

plt.pie(list(org_contributors.values()),autopct='%.2f%%',labels=(list(org_contributors.keys())));
````

Insight：
By this figure, we could see the percentage of contributions made by different organizations

### Types of contributions
By this metric, we could see the number of organizations of different type of contributions.

1. The number of organizations of different type of Issue contributions

````python
df_openEuler=pd.DataFrame(cout_org_datasource) 
draw_bar(df_openEuler, y="issue_org_count", ylabel="Organization Count",title="Organization Count by Issue")
````

2. The number of organizations of different type of Pull Request contributions

````python
df_openEuler=pd.DataFrame(cout_org_datasource) 
draw_bar(df_openEuler, y="pr_org_count", ylabel="Organization Count",title="Organization Count by Pull Request")
````

3. The number of organizations of different type of Git commit contributions

````python
df_openEuler=pd.DataFrame(cout_org_datasource) 
draw_bar(df_openEuler, y="git_org_count", ylabel="Organization Count",title="Organization Count by Git Commit")
````

Insight:

By these figure we could see the growth of organization count  by different data source.
